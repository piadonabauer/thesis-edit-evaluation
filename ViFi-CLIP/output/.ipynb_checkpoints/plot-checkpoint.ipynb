```python
import re
import matplotlib.pyplot as plt
```

```python
# Log-Datei einlesen
#log_file = "few_shot_vitb16_32_2/training_3/log_rank0.txt"
def get_plots(log_file):
    tot_loss = []
    acc_values = {"Acc@1": [], "Acc@5": [], "Acc@10": [], "Acc@50": []}

    with open(log_file, "r") as file:
        for line in file:
            # tot_loss extrahieren
            loss_match = re.search(r"tot_loss ([0-9]+\.[0-9]+)", line)
            if loss_match:
                tot_loss.append(float(loss_match.group(1)))

            # Acc@1, 5, 10, 50 extrahieren
            if log_file=="few_shot_vitb32_128_2/log_rank0.txt":
                acc_match = re.search(r"\* Acc@1 ([0-9]+\.[0-9]+) Acc@5 ([0-9]+\.[0-9]+)", line)
            else:
                acc_match = re.search(r"\* Acc@1 ([0-9]+\.[0-9]+) Acc@5 ([0-9]+\.[0-9]+) Acc@10 ([0-9]+\.[0-9]+) Acc@50 ([0-9]+\.[0-9]+)", line)
            if acc_match:
                acc_values["Acc@1"].append(float(acc_match.group(1)))
                acc_values["Acc@5"].append(float(acc_match.group(2)))
                if not log_file=="few_shot_vitb32_128_2/log_rank0.txt":
                    acc_values["Acc@10"].append(float(acc_match.group(3)))
                    acc_values["Acc@50"].append(float(acc_match.group(4)))

    # Plotten
    epochs = range(len(acc_values["Acc@1"]))  # Annahme: Acc-Werte pro Epoche
    plt.figure(figsize=(12, 6))

    # tot_loss plotten
    plt.subplot(1, 2, 1)
    plt.plot(range(len(tot_loss)), tot_loss, label="tot_loss", color="blue")
    plt.xlabel("Iteration")
    plt.ylabel("Loss")
    plt.title("Training Loss")
    plt.legend()

    # Acc-Werte plotten
    plt.subplot(1, 2, 2)
    for key, values in acc_values.items():
        plt.plot(epochs, values, label=key)
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title("Test Accuracy")
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
get_plots("few_shot_vitb16_32_2/training_3/log_rank0.txt")
    #"few_shot_vitb16_32_2/training_3/log_rank0.txt")
```

## Compare checkpoints

```python
import torch

# Paths to the checkpoints
checkpoint_path_10 = "few_shot_vitb16_32_2/training_2/ckpt_epoch_10.pth"
checkpoint_path_20 = "few_shot_vitb16_32_2/training_2/ckpt_epoch_30.pth"

# Load the checkpoints
checkpoint_10 = torch.load(checkpoint_path_10)
checkpoint_20 = torch.load(checkpoint_path_20)
```

```python
identical = checkpoint_10 == checkpoint_20
print(f"Are the checkpoints identical? {identical}")
```

```python
# Compare model state dictionaries
model_diff = checkpoint_10['model'] == checkpoint_20['model']
print(f"Are the model states identical? {model_diff}")

# Compare optimizer state dictionaries
optimizer_diff = checkpoint_10['optimizer'] == checkpoint_20['optimizer']
print(f"Are the optimizer states identical? {optimizer_diff}")
```

```python
import pickle

# Serialize and compare the state dictionaries
model_10_serialized = pickle.dumps(checkpoint_10['model'])
model_20_serialized = pickle.dumps(checkpoint_20['model'])

identical_model = model_10_serialized == model_20_serialized
print(f"Are the serialized model states identical? {identical_model}")
```

```python
for key in checkpoint_10['model']:
    if torch.equal(checkpoint_10['model'][key], checkpoint_20['model'][key]):
        print(f"{key}: Identical")
    else:
        print(f"{key}: Different")
        print(checkpoint_10['model'][key])
        print(checkpoint_20['model'][key])
```
