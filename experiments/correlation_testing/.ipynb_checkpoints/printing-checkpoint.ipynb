```python
import pandas as pd
import numpy as np
import json
import re
from PIL import Image
import matplotlib.pyplot as plt
```

```python
samples = pd.read_csv("samples.csv", sep=";")
```

```python
samples[:2]
```

```python
with open('edit_turns.json') as f:
    turns = json.load(f)
```

```python
pattern = r'(\d+)-output(\d+)'
```

```python
def show_images_and_prompt(id, turn):
    row = samples[(samples["id"] == id) & (samples["turn"] == turn)]
    
    if row.empty:
        print(f"No matching row found for ID: {id} and Turn: {turn}")
        return

    human_rating_binary = row.iloc[0]["human_rating_binary"]
    if human_rating_binary == 0:
        path = "bad_samples"
    elif human_rating_binary == 1:
        path = "good_samples"
    else: 
        print("Check Binary human rating, neither 0 nor 1!")
        return

    for entry in turns:
        output = entry["output"]
        match = re.search(pattern, output)
        if match:
            found_id = match.group(1)
            found_turn = match.group(2)

            if int(found_id) == id and int(found_turn) == turn:
                input_file = entry["input"]
                mask_file = entry["mask"]
                instruction = entry["instruction"]

                output_image = Image.open(fr'{path}/{output}')
                input_image = Image.open(fr'{path}/{input_file}')
                mask_image = Image.open(fr'{path}/{mask_file}')

                plt.figure(figsize=(15, 5))

                # Output Image
                plt.subplot(1, 3, 1) 
                plt.imshow(output_image)
                plt.title('Output Image')
                plt.axis('off')

                # Input Image
                plt.subplot(1, 3, 2)  
                plt.imshow(input_image)
                plt.title('Input Image')
                plt.axis('off') 

                # Mask Image
                plt.subplot(1, 3, 3)
                plt.imshow(mask_image)
                plt.title('Mask Image')
                plt.axis('off')

                plt.tight_layout()
                plt.show()

                print(f'Instruction: {instruction}')
                return 

    print(f"No matching entry found for ID: {id} and Turn: {turn} in turns.")
```

```python
show_images_and_prompt(id=211860, turn=2)
```

```python
import pandas as pd
import re
import csv
from PIL import Image
import matplotlib.pyplot as plt

# Function to save data to CSV
def save_data_to_csv(file_name, data):
    with open(file_name, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['id', 'turn', 'source_img', 'instruction', 'target_img'])  # Header
        for row in data:
            writer.writerow(row)

def extract_data_to_csv(samples, turns, output_csv="output_data.csv"):
    data = []  # List to hold rows of data

    # Iterate through each row in samples
    for idx, sample in samples.iterrows():
        sample_id = sample["id"]
        turn_number = sample["turn"]
        # Extract matching turn
        for entry in turns:
            output = entry["output"]
            match = re.search(pattern, output)  # Assuming id_turn format
            if match:
                found_id = int(match.group(1))
                found_turn = int(match.group(2))

                if found_id == sample_id and found_turn == turn_number:
                    # Gather the required information
                    source_img = entry["input"]
                    target_img = output
                    instruction = entry["instruction"]
                    # Add this data as a row to the list
                    data.append([sample_id, turn_number, source_img, instruction, target_img])

    # Save the collected data to a CSV file
    save_data_to_csv(output_csv, data)
    print(f"Data successfully saved to {output_csv}")

# Example usage
# samples -> Your samples DataFrame
# turns -> Your list of entries from the dataset
#extract_data_to_csv(samples, turns)
```

```python
!pip install pyarrow
import pandas as pd
import os
from PIL import Image
import io
```

```python


# Paths for the image directories
parquet_dir = '/home/jovyan/BA/Github/MagicBrush/data/'
base_image_dir = '/home/jovyan/BA/Github/MagicBrush/'

# Create a directory for images if it doesn't exist
os.makedirs(base_image_dir, exist_ok=True)

# Initialize DataFrames
train_df = pd.DataFrame(columns=['img_id', 'turn_index', 'source_img', 'target_img', 'instruction'])
dev_df = pd.DataFrame(columns=['img_id', 'turn_index', 'source_img', 'target_img', 'instruction'])

def process_image(image, save_dir):
    image_data = image['bytes']
    img = Image.open(io.BytesIO(image_data))
    img = img.resize((512, 512))

    save_path = os.path.join(save_dir, image['path'])
    img.save(save_path)
    return save_path

# Loop over all parquet files
for filename in os.listdir(parquet_dir):
    if filename.endswith(".parquet"):
        file_path = os.path.join(parquet_dir, filename)
        temp_df = pd.read_parquet(file_path)
        
        # Create a directory for the images related to this parquet file
        parquet_image_dir = os.path.join(base_image_dir, os.path.splitext(filename)[0])
        os.makedirs(parquet_image_dir, exist_ok=True)
        
        # Filter out the mask_img column if it exists
        if 'mask_img' in temp_df.columns:
            temp_df = temp_df.drop(columns=['mask_img'])

        # Prepare a temporary dataframe to store the necessary columns
        temp_processed_df = pd.DataFrame(columns=['img_id', 'turn_index', 'source_img', 'target_img', 'instruction'])

        # Process train files
        if filename.startswith("train"):
            # Process each row
            processed_rows = []
            for idx, row in temp_df.iterrows():
                if 'img_id' in row and 'turn_index' in row and 'source_img' in row and 'target_img' in row and 'instruction' in row:

                    source_path = process_image(row['source_img'], parquet_image_dir)
                    source_img_path = source_path
                    
                    # Process target image
                    target_path = process_image(row['target_img'], parquet_image_dir)
                    target_img_path = target_path

                    # Create a new row with the required columns
                    new_row = {
                        'img_id': row['img_id'],
                        'turn_index': row['turn_index'],
                        'source_img': source_img_path,
                        'target_img': target_img_path,
                        'instruction': row['instruction']
                    }
                    processed_rows.append(new_row)

            temp_processed_df = pd.concat([temp_processed_df, pd.DataFrame(processed_rows)], ignore_index=True)

            # Append to the train dataframe
            train_df = pd.concat([train_df, temp_processed_df])

        # Process dev files
        elif filename.startswith("dev"):
            processed_rows = []
            for idx, row in temp_df.iterrows():
                if 'img_id' in row and 'turn_index' in row and 'source_img' in row and 'target_img' in row and 'instruction' in row:

                    source_path = process_image(row['source_img'], parquet_image_dir)
                    source_img_path = source_path 
                    
                    # Process target image
                    target_path = process_image(row['target_img'], parquet_image_dir)
                    target_img_path = target_path 
                    
                    # Create a new row with the required columns
                    new_row = {
                        'img_id': row['img_id'],
                        'turn_index': row['turn_index'],
                        'source_img': source_img_path,
                        'target_img': target_img_path,
                        'instruction': row['instruction']
                    }
                    processed_rows.append(new_row)

            temp_processed_df = pd.concat([temp_processed_df, pd.DataFrame(processed_rows)], ignore_index=True)

            # Append to the dev dataframe
            dev_df = pd.concat([dev_df, temp_processed_df])

# Save the updated DataFrames to CSV
train_df.to_csv(os.path.join(base_image_dir, 'train_data.csv'), index=False)
dev_df.to_csv(os.path.join(base_image_dir, 'dev_data.csv'), index=False)

print("Images processed and DataFrames saved successfully!")
```
